```python
import pandas as pd
import nepali_roman as nr
```

```python
df1= pd.read_csv('../Data/ss_ac_at_txt_unbal.csv')
df1.head()
```

```python
df1.columns
```

```python
df2 = pd.DataFrame(columns=['Polarity', 'Sentiment', 'Aspect Term', 'Text'])
df2.head()
```

```python
df2.loc[len(df1)] = df1.columns
df2 = df2.reset_index(drop=True)

df2.head()
```

```python
df1= df1.rename(columns={'0':'Polarity', 'GENERAL':'Sentiment', 'जोगाउन को लागि':'Aspect Term', 'गुठी विधेक ल्याएर ठमेल मा राज गुठि को जग्गा मा बने को छाया सेन्टर जस्ता लाई जोगाउन को लागि ल्याउदैछ विधेक ।':'Text'})
df1.head()
```

```python
df= pd.concat([df2,df1], ignore_index=True)
df.head()
```

```python

# Reorder columns
df = df[['Text', 'Aspect Term', 'Sentiment','Polarity']]
df.head()
```

```python
# Saving this dataset with after reordering columns
```

```python
df.to_csv('../Data/Nepali_dataset.csv', index= False)
```

# Making a translated Dataset


```python
df_translated = df.copy()
df_translated.head()
```

```python
# Applying romanize_text to the 'Text' and 'Aspect Term' columns
df_translated['Text'] = df_translated['Text'].apply(nr.romanize_text)
df_translated['Aspect Term'] = df_translated['Aspect Term'].apply(nr.romanize_text)

df_translated.head()
```

```python
df_translated.to_csv('../Data/Nepali_dataset_romanized.csv',index=False)
```
