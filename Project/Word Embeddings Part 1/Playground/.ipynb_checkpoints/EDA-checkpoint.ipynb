# Checking vocabulary

```python

texts = df_copy['Text']

# Tokenizing and splitting into words
all_words = " ".join(texts).split()

# Count unique words
unique_words = set(all_words)

# Vocabulary size
vocabulary_size = len(unique_words)
print(f"Vocabulary Size: {vocabulary_size}")


```

```python
from collections import Counter

# Count word frequencies
word_counts = Counter(all_words)

# Get the 10 most common words
most_common_words = word_counts.most_common(10)

# Get the 10 least common words
least_common_words = word_counts.most_common()[:-11:-1]  # last 10 elements (reverse order)

# Print top 10 most common words
print("Top 10 most frequent words:")
for word, freq in most_common_words:
    print(f"{word}: {freq}")

# Print top 10 least common words
print("\nTop 10 least frequent words:")
for word, freq in least_common_words:
    print(f"{word}: {freq}")
```

```python

from wordcloud import WordCloud

def generate_wordcloud_for_class(df, class_name):
    """
    Generate a WordCloud for a given class from the DataFrame.

    Args:
    - df (pd.DataFrame): The input DataFrame containing 'Text' and 'Class' columns.
    - class_name (str): The class for which to generate the WordCloud (e.g., 'GENERAL' or 'VIOLENCE').

    Returns:
    - None
    """
    # Filter the DataFrame for the specific class
    filtered_df = df[df['Sentiment'] == class_name]
    words = " ".join(filtered_df['Text'])
    
    # Font path for Devanagari script as wordCloud requires to specify font for such data
    font_path = '/usr/share/fonts/truetype/lohit-devanagari/Lohit-Devanagari.ttf'
    
    # Generate WordCloud
    wordcloud = WordCloud(
        width=800, 
        height=400, 
        background_color='white', 
        min_font_size=10, 
        collocations=False, 
        normalize_plurals=True,
        font_path=font_path
    ).generate(words)
    
    # Display the WordCloud
    plt.figure(figsize=(10, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')  # Turn off axis labels
    plt.title(f'WordCloud for class: {class_name}')
    plt.show()


```

# N-Gram Analysis 

```python
def test_n_grams(text, n_gram=1):
    print("The text before n-grams:", text)
    # Split the text into words
    words = text.split()
    print("Words before n-grams:", words)
    
    # Generate n-grams from the list of words
    temp = zip(*[words[i:] for i in range(0, n_gram)])
    ans = [' '.join(n_gram) for n_gram in temp]
    return ans

# Testing with sample text
text = 'गुठी विधेक ल्याएर ठमेल मा राज गुठि को जग्गा मा बने को छाया सेन्टर जस्ता लाई जोगाउन को लागि ल्याउदैछ विधेक ।'
print(test_n_grams(text, 3))
```

```python
# Function to create n_grams for each instances
def add_n_grams_column(dataframe, column_name, n_gram=1):
    """
    Generate n_grams by passing the dataframe, the name of the column that consists the text, n-grams to generate
    dataframe= pass pandas dataframe
    column_name= passs the column that consists the corpus
    n_gram= n_grams to generate, default 1
    top_n= returns top 10 frequently occuring n-gram in the corpus, default 10
    
    the function will create n-grams array for each instances, so it's advisable to pass a copy of the original dataset
    """
    # New column name
    n_gram_column_name = f'n-gram-{n_gram}'
    
    # Function to generate n-grams for a single text
    def generate_n_grams(text):
        if not isinstance(text, str):  # Handle NaN or non-string data
            return []
        words = text.split()
        temp = zip(*[words[i:] for i in range(n_gram)])
        return [' '.join(n_gram) for n_gram in temp]
    
    # Applying the function to each row in the column
    dataframe[n_gram_column_name] = dataframe[column_name].apply(generate_n_grams)

add_n_grams_column(df_copy, column_name='Text', n_gram=3)
df_copy.head()
```

```python
# Now checking the top n_grams through the new column

def get_top_n_grams(dataframe, n_gram_column, top_n=10):
    """
    Get the top n most frequent n-grams from a specific column.
    
    :dataframe: pandas DataFrame containing the n-grams column
    :n_gram_column: Name of the column containing n-grams
    :top_n: Number of top frequent n-grams to display, default 10
    :return: DataFrame with top n-grams and their counts
    """
    
    # Getting all the n-grams and count frequencies
    all_n_grams = [n_gram for n_grams in dataframe[n_gram_column] for n_gram in n_grams]
    n_gram_counts = Counter(all_n_grams)
    
    # Converting to a DataFrame
    most_common = n_gram_counts.most_common(top_n)
    return pd.DataFrame(most_common, columns=[ n_gram_column, 'count'])

# The count is visualized through a dataframe returned from function get_top_n_grams
n_gram_df= get_top_n_grams(df_copy, 'n-gram-3')
n_gram_df.head(10)
```

## TF-IDF
 it's a numerical statistic that measures how important a word is in a document relative to a collection of documents.

```python
# making a corpus of text data
corpus = df_copy['Text'].to_numpy()
print(corpus)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer= TfidfVectorizer()
vectorizer_model = vectorizer.fit_transform(corpus)

# get idf values
print('\nidf values:')
for ele1, ele2 in zip(vectorizer.get_feature_names_out(), vectorizer.idf_):
    print(ele1, ':', ele2)


```

```python
# the vectorizer is misinterpreting the Nepali scripts, so let's ensure proper tokenization as suggested by chatGPT
import re

# Define a custom tokenizer for Nepali text
def nepali_tokenizer(text):
    # Remove unwanted characters and tokenize
    tokens = re.findall(r'[\u0900-\u097F]+', text)  # Match Devanagari script
    return tokens
    
# Use the custom tokenizer in TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=nepali_tokenizer)

# Fit and transform the corpus
vectorizer_model = vectorizer.fit_transform(corpus)

# Get idf values
print('\nidf values:')
for word, idf in zip(vectorizer.get_feature_names_out(), vectorizer.idf_):
    print(word, ':', idf)

```

```python
# get indexing
print('\nWord indexes:')
print(vectorizer.vocabulary_)

```

```python
# display tf-idf values
print('\ntf-idf value:')
print(vectorizer_model)

```

```python

# in matrix form
print('\ntf-idf values in matrix form:')
print(vectorizer_model.toarray())
```

```python
# Convert to a DataFrame
df_tfidf = pd.DataFrame(vectorizer_model.toarray(), columns=vectorizer.get_feature_names_out())

# Display summary and 
print("TF-IDF Matrix (Shape):", df_tfidf.shape)
print("\nTop terms by mean TF-IDF score:")
print(df_tfidf.mean().sort_values(ascending=False).head(20))
```

The top mean tf-idf scores suggests words that might carry context, especially for instances of targeted hate. While we can observe some words like सर, पर्छ, छ, ।, again Stop words removal seem inticing

```python
import matplotlib

# Set the font to use Lohit Devanagari
matplotlib.rcParams['font.family'] = 'Lohit Devanagari'


# Heatmap generation function for tf-idf
def heatmap_tfidf(dataframe, documents=10, terms=10):
    """
    - Pass Dataframe created from arrays of tf_idf vectorized model and the column names
    - documents: number of document/text instances to visualize, default 10
    - terms: no of terms to visualize. default 10
    since random terms won't give us much information--> making the function show high mean tf-idf terms
    """
    
    plt.figure(figsize=(12, 8))
    
    sorted_terms = dataframe.mean().sort_values(ascending=False).index
 
    sns.heatmap(dataframe[sorted_terms[:terms]].iloc[:documents], cmap="coolwarm", annot=True, fmt=".2f")
    plt.title("TF-IDF Heatmap (Subset)")
    plt.xlabel("Terms")
    plt.ylabel("Documents")
    plt.xticks(rotation=45, ha='right')
    
    plt.show()

heatmap_tfidf(df_tfidf, terms= 30)
```
